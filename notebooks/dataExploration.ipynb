Correlation

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize = ( 12, 10))
sns.heatmap(data.corr(),annot = True, linewidth = 0.5, fmt = '.1f', ax = ax)
```

The plot shows correlation between radius, perimeter, area. More correlation for _mean and _worst values.
Also texture_worst and texture_mean, fractal_dimension_worst and fractal_dimension_mean.
